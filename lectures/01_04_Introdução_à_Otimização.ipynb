# Otimização

*Almost everything is optimization ... or needs optimization ...* (Xin-She Yang)

---

* Muitos problemas importantes podem ser formulados como problemas de otimização
 * Maximizar eficiência, precisão, lucro, desempenho, sustentabilidade ...
 * Minimizar custos, consumo de energia, desperdício, distância viajada, impacto ambiental...
 * São o ”core” de muitas empresas (tripadvisor, waze...)
* Resolver problemas de otimização é desafiador
 * **Reduzir** um problema novo a uma instância de outro bem conhecido permite o uso de soluções pré-existentes
 * Algoritmos gulosos/gananciosos são uma forma prática de encontrar uma solução aproximada para um problema de otimização


## O que é um modelo de otimização?

* Uma função objetivo que deve ser maximizada ou minimizada, ex:
 * Minimizar o tempo de viagem entre Recife e Belém
 * Maximizar o número de clientes atendidos em uma loja
* Um conjunto de restrições que devem ser cumpridas, ex:
 * Não quero gastar mais do que $100
 * Devo estar em Belém às 20h
 * Um atendente não pode ficar mais de 20% do tempo ocioso

## Componentes de um problema de otimização

**Variáveis do problema**
* Inteiras, binárias, contínuas...

**Função Objetivo**
* Função sobre as variáveis do problema (Linear / Não-linear)
* Pode haver mais de uma (mono vs multi-objetivo)
* Minimizar ou Maximizar

**Restrições**
* Inequações sobre as variáveis do problema
* Equações sobre as variáveis do problema

**Domínio das variáveis do problema**
* Indicam o espaço de busca sobre a variável

# Modelagem de problemas de otimização

## Forma geral de um problema de otimização

O problema pode possui uma ou mais funções objetivo (representadas abaixo de forma vetorial)

$f(\pmb{x}) = [f_1(\pmb{x}),f_2(\pmb{x}),\ldots,f_m(\pmb{x})]$

Cada função pode ter como fim minimizar ($\min$) ou maximizar ($\max$). Caso $m=1$, temos um problema monoobjetivo.

O problema ainda pode ser irrestrito ou restrito, i.e., estar sujeito à múltiplas restrições, que podem ser equações ou inequações:

$h_i(\pmb{x}) = 0$, $(i=1,2,\ldots,M)$,

$g_j(\pmb{x}) \leq 0$, $(j=1,2,\ldots,N)$

Além disso, cada variável pertencente a um domínio:

$\pmb{x} = (x_1,x_2,\ldots,x_D) \in \mathbb{R}^D$

O domínio poderia ser ainda:
 * contínuo ou inteiro
 * limitado ou ilimitado

## Subtipos

* A área de otimização é chamada de programação matemática
* Diversas subáreas
  * Programação linear
    * Estuda problemas onde a função objetivo e as restrições são lineares
  * Programação inteira
    * Estuda problemas lineares com uma ou mais variáveis inteiras
  * Otimização combinatória
    * Estuda problemas lineares ou não lineares onde o espaço de soluções é discreto e limitado
  * Programação não-linear
    * Estuda o caso geral quando a função objetivo ou as restrições contém partes não-lineares

# Solução de problemas de otimização

* Como solucionar um problema de otimização?
 * Depende da complexidade do problema
 * Depende dos tipos das variáveis
 * Depende das funções
* Métodos analíticos
  * Sem restrições
    * Método da derivada
  * Com restrições
    * Relaxação de Lagrange + método da derivada
* Métodos Numéricos
  * Baseados na execução de um algoritmo


## Métodos Numéricos (Algoritmos)

* Essência de um Algoritmo de Otimização
  * Gera uma solução melhor $\pmb{x}^{(t+1)}$ a partir de uma solução existente $\pmb{x}^{(t)}$, tal que $\pmb{x}^{(t+1)} = A(\pmb{x}^{(t)},\alpha)$, onde $\alpha$ é um conjunto de parâmetros e $A$ é uma estratégia de atualização da solução usada pelo algoritmo
  * Algoritmos baseados em populações usam múltiplos caminhos
  * Algoritmos diferentes $\Rightarrow$ formas diferentes de gerar soluções
* Exatos (Convencionais)
  * Simplex (PL)
  * Busca Exaustiva (força bruta)
  * Branch-and-Bound (PLI)
  * Método de Newton (PNL)
* Heurísticas: "seek <u>good feasible solutions</u> to optimization problems in circumstances where <u>the complexity of the problem</u> or the <u>limited time available</u> for its solution <u>do not allow exact solution</u>" [1]
  * Heurísticas específicas para o problema em si
    * Aproximações
  * Meta-heurísticas
    * Algoritmos Genéticos, Simulated Annealing, PSO...



[1] [Rardin, R.L. and Uzsoy, R., 2001. Experimental evaluation of heuristic optimization algorithms: A tutorial. Journal of Heuristics, 7, pp.261-304.](https://link.springer.com/article/10.1023/A:1011319115230)

### Algoritmos Convencionais vs Algoritmos Bio-inspirados

* Algoritmos convencionais
  * A otimalidade global (são **buscas locais**) depende do problema
    * A solução de problemas lineares ou não-lineares convexos é global, mas outros problemas não possuem esta garantia
  * Os resultados dependem dos pontos iniciais (exceto para problemas lineares e convexos)
  * Tem dificuldade com discontinuidades na função objetivo
* Algoritmos bio-inspirados
  * Aumentam as chances de encontrar soluções globais
  * Resolvem uma grande classe de problemas (são como caixas pretas)
  * Podem ser potencialmente mais caros computacionalmente

# Exemplo

Certa empresa fabrica 2 produtos P1 e P2. O lucro por unidade de P1 é de \$100 e o lucro unitário de P2 é de \$150. A empresa necessita de 2 horas para fabricar uma unidade de P1 e 3 horas para fabricar uma unidade de P2. O tempo mensal disponível para essas atividades é de 120 horas. A demanda esperada para P1 por mês é de 40 unidades e para P2 é de 30 unidades por mês. Quanto a empresa deveria fabricar de cada produto mensalmente de modo a maximizar seu lucro?

# Modelo

$P_1$ é a quantidade de produtos do tipo P1 produzidos

$P_2$ é a quantidade de produtos do tipo P2 produzidos

$
\max 100P_1 + 150P_2 \\
s.t. \\
2P_1 + 3P_2 \leq 120 \\
P_1 \leq 40 \\
P_2 \leq 30
$

#Solução

## Instalando a biblioteca de modelagem

```python
!pip install -q pyomo
```

## Transcrevendo o modelo

```python
from pyomo.environ import *

# create a model
model = ConcreteModel()

# declare decision variables
model.p1 = Var(domain=NonNegativeIntegers)
model.p2 = Var(domain=NonNegativeIntegers)

# declare objective
model.lucro = Objective(expr = 100*model.p1 + 150*model.p2, sense=maximize)

# declare constraints
model.horas = Constraint(expr = 2*model.p1 + 3*model.p2 <= 120)
model.demandaP1 = Constraint(expr = model.p1 <= 40)
model.demandaP2 = Constraint(expr = model.p2 <= 30)

model.pprint()
```

## Instalando o solver

```python
!apt-get install -y -qq glpk-utils
```

## Solucionando

```python
SolverFactory('glpk', executable='/usr/bin/glpsol').solve(model).write()

# display solution
print('\nLucro = ', model.lucro())

print('\nDecision Variables')
print('P1 = ', model.p1())
print('P2 = ', model.p2())

print('\nConstraints')
print('Horas  = ', model.horas())
print('Demanda P1 = ', model.demandaP1())
print('Demanda P2 = ', model.demandaP2())
```
