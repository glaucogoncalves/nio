## Árvore geradora mínima

Dado que:
* Árvore é um grafo sem ciclos
* Uma <u>árvore geradora mínima</u> (*minimum spanning tree*) é um subgrafo que possui todos os nós do grafo original e cuja soma dos pesos é mínima
* Grafo $G=(N,E)$ com cada aresta tendo um peso $w_{i,j}$
* $n$ é o número de nós do grafo
* $x_{ij}$ indica se uma aresta do grafo está sendo usada na árvore

---

$
\min \sum_{i}\sum_{j} w_{ij}x_{ij} \\
\text{s.t.} \\
\sum_i\sum_j x_{ij} = n-1, \; \forall i,j \in E \\
\sum_i\sum_j x_{ij} \leq |S|-1, \; \forall i,j \in S, \; \forall S \subset E\\
x_{ij} \in {0,1}, \; \forall i,j \in E
$

image.png

###Solução

1. Simplex $O(2^n)$
2. Algoritmo de Kruskal
3. Algoritmo de Prim

### Algorito de Prim

---
**Entradas**

$G = (N,E)$: um grafo com cada aresta $(u,v)$ tendo peso $w_{u,v}$ \\

---

**Saídas**

$T = (N,S)$: Uma árvore com um conjunto de arestas $S \subset E$ tal que $\sum_{i \in S} w_i$ é mínimo \\

---

**Algoritmo**
1. Inicialize a árvore $T$ com um  nó $u \in N$
2. Repita:
3. $\;$ Escolha a aresta (u,v) em $E$ com o menor $w_{u,v}$ que se ligue a algum nó que já está em $T$
4. $\;$ Adicione o nó que está na outra ponta da aresta à $T$
5. $\;$ Adicione a aresta à $T$
6. Até que todos os nós de $G$ sejam adicionados à $T$

---

##Problema do Caminho mais Curto

Dado que:
* Grafo $G=(N,E)$ com cada aresta tendo um peso $w_{i,j}$
* $O$ é a origem e $T$ é o destino
* $x_{ij}$ indica se uma aresta do grafo está sendo usada no caminho

---



$
\min \sum_{i}\sum_{j} w_{ij}x_{ij} \\
\text{s.t.} \\
\sum_j x_{ij} =, \; \text{se} \; i = O \\
\sum_k x_{ki} = 1, \; \text{se} \; i = T \\
\sum_k x_{ki} - \sum_j x_{ij}  = 0, \; \text{se} \; i \neq O,T \\
x_{ij} \in {0,1}, \; \forall i,j \in E
$

###Solução

1. Simplex $O(2^n)$
2. Algoritmo de Djikstra

### Algoritmo de Djikstra

**Observação**: O algoritmo resolve o problema da árvore de menores caminhos, que é um problema ligeiramente diferente. Contudo o mesmo algoritmo pode ser usado para o problema acima, bastando parar o laço das linhas 6-11 quando encontrar o nó de destino ($T$).

---
**Entradas**

$G = (N,E)$: um grafo com cada aresta $(u,v)$ tendo peso $w_{u,v}$ \\
$u$: Um nó inicial tal que $u \in N$

---

**Saídas**

$T = (N,S)$: Uma árvore com um conjunto de arestas $S \subset E$ tal que, $\forall v \in N$, a distância $d(u,v)$ é mínima \\

---

**Algoritmo**
1. Para cada nó $v \in N$, faça:
2. $\;$ Se os nós $u$ e $v$ estão conectados em $G$
3. $\;$ Então incialize a distância $d(u,v) = w_{u,v}$
4. $\;$ Senão incialize a distância $d(u,v) = \infty$
5. Inicialize a árvore $T$ com o nó $u$
6. Repita
7. $\;$ Selecione o nó $v$, ainda não pertecente à árvore, cuja distância $d(u,v)$ é a menor
8. $\;$ Adicione à $T$ o nó $v$ e a aresta que liga $u$ a $v$
9. $\;$ Para cada nó $s$ conectado a $v$, faça:
10. $\;$$\;$$\;$ $d(u,s) = min(d(u,s),d(u,v)+w_{v,s})$
11. Até que todos os nós de $G$ sejam adicionados à $T$

## Problema Clássico de Transporte



image.png

$c_{ij}$ = custo unitário de transporte do fornecedor i (i = 1,…, m) para o consumidor j (j = 1,…, n)

$Cf_i$ = capacidade de abastecimento do fornecedor i (i = 1,…, m)

$d_j$ = demanda do consumidor j (j = 1,…, n)

Variáveis de decisão:

$x_{ij}$ = quantidades transportadas do fornecedor i (i = 1,…, m) para o consumidor j (j = 1,…, n)

---

$
\min \sum_{i}\sum_{j} c_{ij}x_{ij} \\
\text{s.t.} \\
\sum_j x_{ij} \leq Cf_i, \; i=1,\ldots,m \\
\sum_i x_{ij} \geq d_j, \; j=1,\ldots,n \\
x_{ij} \geq 0, \; \forall i,j
$

###Variações
1. Balanceado
2. Fornecimento Total é Maior do que Demanda Total
3. Capacidade Total de Fornecimento é Menor do que Demanda Total Consumida

Os casos 2) e 3) podem ser convertidos no caso 1):
* Se estivermos no caso 2) criamos um consumidor extra cuja demanda $d_{extra}$ é dada pela diferença $d_{extra} = \sum_i Cf_i - \sum_j d_j$
* Se estivermos no caso 3) criamos um fornecedor extra cuja demanda capacidade de abastecimento $Cf_{extra}$ é dada pela diferença $Cf_{extra} = \sum_j d_j - \sum_i Cf_i$

###Solução

1. Simplex $O(2^n)$
2. O Algoritmo de Transporte

### Algoritmo de Transporte

---
**Condições**

* O problema deve estar balanceado.

---
**Entradas**

* Os custos de transporte entre cada fornecedor e consumidor $c_ij$, a capacidade de abastecimento de cada fornecedor $Cf_i$ e a demanda de cada consumidor $d_j$. \\
* O problema deve ser colocado na forma tabular abaixo.

image.png

---

**Saídas**

$x_{ij}$: a quantidade transportada do fornecedor $i$ para o consumidor $j$.


**Algoritmo**
1. Encontre uma solução básica factível (SBF) inicial. Abaixo usamos o <u>método do canto noroeste</u> (existem outros)
 * Faça $C^R_i = Cf_i$, onde $C^R_i$ é a capacidade restante do fornecedor $i$
 * Faça $D^R_j = d_j$, onde $D^R_j$ é a demanda restante do consumidor $j$
 * Faça $x_{ij} = \text{None} \; \forall i,j$
 * Para $i =  1 \ldots n$ faça **#começamos pela célula no canto superior esquerdo (noroeste)**
    * Para $j =  1 \ldots m$ faça
      * Se $x_{ij} == \text{None}$ então
         * $x_{ij} = D^R_j$ **#Aloque a maior quantidade possível de produto a célula**
         * Para $i' = i+1 \ldots n$ faça $x_{i'j} = 0$ **#Marque com 0 as células correspondentes a mesma linha**
      * Senão
         * $x_{ij} = C^R_i$
         * Para $j' = j+1 \ldots n$ faça $x_{ij'} = 0$
      * $D^R_j = D^R_j - x_{ij}$
      * $C^R_j = C^R_j - x_{ij}$
2. Verifique se a SBF é ótima
  * Determinar os valores de $u_i$ e $v_j$ tal que $u_i+v_j = c_{ij}$,  para cada <u>variável básica</u> ($x_{ij} \neq 0$)
    * Escolhe-se $u_1 = 0$, logo $v_j = c_{1j}$ e daí segue-se definindo os demais $u_i$ e $v_j$
  * Para as variáveis não-básicas determina-se o custo reduzido $\hat{c_{ij}} = ui + vj - cij$
  * Se algum $\hat{c_{ij}} \gt 0$ então não atingimos o ótimo, caso contrário paramos e a SBF é ótima
3. Se a SBF não é ótima, então gere uma nova SBF:
  * Escolha a variável não básica de maior $\hat{c_{ij}}$ para ser básica
  * Um ciclo fechado deve ser construído para determinar a variável que sairá da base. Esse ciclo fechado deve:
    * Iniciar e finalizar na variável não básica escolhida
    * Ser formado por uma sequência de segmentos horizontais e verticais conectados entre si
    * Cada esquina do ciclo deve ser uma variável básica, exceto a variável escolhida
  * Escolhemos o menor valor dentre células vizinhas da célula da variável não-básica
    * Este valor será subtraído das células vizinhas
    * Este valor será somado às outras duas células

5. Vá para 2)
