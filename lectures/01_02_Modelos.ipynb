# A MODELAGEM E A ENGENHARIA

## A importância da abstração e da modelagem

Conforme as Diretrizes Curriculares Nacionais para os Cursos de Engenharia, o perfil do egresso dos cursos de Engenharia é um profissional que deve, dentre outras características, “ser capaz de reconhecer as necessidades dos usuários, formular, analisar e resolver, de forma criativa, os problemas de Engenharia”. Para atingir esta meta o egresso deve cercar-se de todo o ferramental científico, tecnológico e metodológico para “ser capaz de modelar os fenômenos, os sistemas físicos e químicos, utilizando as ferramentas matemáticas, estatísticas, computacionais e de simulação, entre outras.”.

Nestas linhas gerais, pode-se perguntar qual a principal habilidade a ser exercitada ao longo de um curso de Engenharia? Certamente, responder esta pergunta não é uma tarefa fácil. Cada professor de um curso de Engenharia terá uma lista com diversas respostas, as quais dependerão de sua área de atuação, experiência e visão de mundo. Em nossa opinião, no entanto, a **abstração** é uma importante habilidade que um egresso deve possuir.

Pode-se dizer que a abstração é uma operação mental pela qual se compreende um fenômeno ou conceito através da consideração de somente algumas de suas características. No caso do profissional de engenharia, esta operação é aplicada nas principais atividades de seu dia-a-dia, a saber: o projeto, adaptação e melhoria de sistemas de sua área de atuação. Para alcançar a real melhoria de um sistema, o profissional deve considerá-lo como um todo, focando porém nas características do sistema que realmente importam. Este tipo de olhar demanda justamente a abstração, que é a habilidade-chave que permite realizar bem estas atividades.

Computação e engenharias trabalham com modelos, podendo ser modelos algébricos, probabilísticos, estatísticos, modelos computacionais de eventos discretos, modelos processuais etc. Um modelo é uma representação de fenômenos (eventos) ou sistemas (coisas) reais ou artificiais. A criação de um modelo envolve justamente a abstração, que será necessária para bem selecionar os aspectos mais importantes e fornecer uma representação confiável do que se está analisando. Do ponto de vista de sistemas, a **finalidade** de um modelo é permitir a análise das características do sistema a qual pode ser usada para a avaliação do desempenho esperado ou para a otimização do funcionamento. Na avaliação de desempenho podem-se identificar quais as variáveis mais importantes no sistema e onde se encontram os gargalos do mesmo. Na otimização, pode-se realizar mudanças nas variáveis do modelo para obter redução de um gargalo ou melhoria do desempenho de forma geral.

Uma das principais vantagens na resolução de problemas por meio de modelos é que um modelo pode ser reusado facilmente. Assim, após construir um modelo de um sistema, o dono do modelo pode testá-lo em diferentes situações e inferir o comportamento do sistema diante daquelas situações. Desta forma, é possível inferir o comportamento do sistema sem perturbá-lo. Pode-se ainda descobrir o comportamento do modelo frente a situações não previstas na concepção do sistema.


## Conceitos gerais de modelagem

A tarefa de resolver problemas (desde sua modelagem até sua solução final) é pouco exercitada nos cursos. Geralmente trabalha-se em uma abordagem *bottom-up* em que as ferramentas são ensinadas sem qualquer ligação com problemas reais.

Esta abordagem fica evidenciada na estrutura dos currículos de engenharia. A formação básica e tradicional de um engenheiro envolve a aprendizagem de ferramental matemático (Cálculo, Álgebra Linear, Probabilidade etc) e conceitual (Física, Química etc) ao longo dos dois primeiros anos de estudo. Só após este período o estudante tem acesso a um conjunto de ferramentas profissionais que “aplicam” parte do ferramental matemático e parte dos conceitos aprendidos em problemas bem definidos previamente escolhidos para avaliação. O resultado deste modelo é que **aprende-se apenas a resolver bem os exercícios do capítulo**.
  
Um dos motivos da manutenção desta abordagem na maioria dos cursos de engenharia é a demanda de energia necessária por parte de discentes e docentes. Observe-se que a solução de problemas reais passa por um conjunto de etapas, as quais são condensadas no **Procedimento Geral de Modelagem** mostrado abaixo. A energia para conduzir todos os estudantes ao longo de todo este processo em um ou mais problemas (geralmente mais de um, para que se cubra diferentes aspectos da disciplina em estudo) é alta. Enfim, apresentar conteúdos por meio de problemas práticos e reais requer uma mudança profunda na forma como estudantes e professor se relacionam nas disciplinas, exigindo maior dedicação de ambos na condução das discussões sobre os problemas e suas soluções.

## Sobre a definição de um problema

A respeito do Procedimento Geral de Modelagem deve-se observar que, no mundo real, todo o processo se inicia a partir do problema e não do problema definido. Este é um ponto fundamental na modelagem e muitas vezes negligenciado. Normalmente um livro didático contém centenas de problemas bem definidos, mas nunca um problema mal definido ou indefinido. Contudo, problemas reais não se encontram declarados e então a sua definição por meio de aspectos gerais como escopo, objetivo, justificativas e outros são essenciais para que o problema possa ser estudado e sua solução encontrada.

A etapa de declaração do problema, se mal conduzida, pode levar a um problema mal-definido. Veja, por exemplo, as três formas da Figura 2. Um problema mal definido pode ser simplesmente enunciado como: “forme grupos com essas formas”. Pare um pouco sua leitura e pense acerca deste problema, continuamos logo abaixo.


**Agrupe estas formas**

image.png


Certamente sua reação ao problema deve ter sido a de questionar: “mas agrupar como”? O fato de que o problema não contém qualquer outra especificação acerca de como os grupos serão formados, faz com que ele admita uma grande série de soluções diferentes. Procure agora imaginar uma forma de deixar este problema mais claro, para isso procure acrescentar algum qualificativo ao problema, por exemplo: “forme grupos com essas formas, tal que figuras com ângulos retos fiquem no mesmo grupo”.

Este exemplo simples procura ilustrar o que é um problema mal definido. Contudo, o processo de declarar problemas a partir da realidade é mais complexo e, pode-se dizer, é a primeira atividade a exigir abstração da parte de quem esteja tratando o problema.

## A Modelagem

Tendo um problema definido, passa-se à etapa de modelagem. O modelo é uma codificação do problema em algum tipo de linguagem (visual, matemática, etc) que permita o entendimento do problema de forma clara e inequívoca, bem como, em alguns casos, solução por métodos quantitativos.

Um mapa é um modelo. Ele representa uma região, mas apresenta somente alguns aspectos de maior interesse daquela região. Na figura abaixo apresentamos um mapa (da plataforma Google Maps) de uma parte do município de Belém. Nele podemos ver alguns logradouros de Belém (Complexo Ver-o-Rio, Universidade Federal do Pará, Hospital Gaspar Vianna etc), bem como os nomes de certos bairros (Umarizal, Campina, Marco etc). Podemos ainda ver as principais vias (avenidas, ruas etc) e áreas verdes. Obviamente, o mapa não reflete toda a realidade de Belém, como cada um de seus prédios ou mesmo de suas ruas.

Contudo novas informações podem ser acrescentadas ao mapa de Belém, como exemplo temos o mapa na figura (b) que acrescenta o estado do trânsito nas principais vias de Belém em 10 de Agosto de 2022 às 15 horas. Esta informação traz novos dados ao modelo, porém deve-se observar que ele acrescenta complexidade ao modelo, tornando-o mais difícil de ler. Retomaremos este aspecto mais à frente, mas aqui fica claro que a modelagem deve estabelecer um compromisso entre a complexidade do modelo e sua confiabilidade em representar o problema de interesse.

A modelagem de um problema deve levar diversos aspectos em consideração. Por exemplo, deve-se escolher que tipo de modelo é mais adequado ao problema. Podemos usar um modelo baseado na física (propagação de sinal, movimento, termodinâmica etc) ou baseado na aplicação de teorias não-físicas para explicação do fenômeno. Podemos ainda decidir se o tempo é uma variável de importância, e se deve ser tratado de forma contínua ou discreta.
Um modelo é chamado de Caixa Branca quando conhecemos as regras (equações) que regem este modelo, as quais são usadas para apresentar o modelo e solucioná-lo. Um modelo Caixa Preta, por outro lado, considera apenas as entradas e saídas do problema, sem qualquer conhecimento de seus aspectos internos. É comum nestes casos a elaboração de funções de transferência obtidas empiricamente. Deve-se ainda observar que um modelo pode ter uma mistura destes, o qual é denominado como modelo Caixa Cinza.

Independentemente do tipo de modelo, devemos considerar com cuidado quais características são realmente importantes em um modelo. A Figura 4 procura ilustrar este conceito. Nela vemos um problema P em algum ponto t do tempo. Este problema apresenta diversas características (os pontos pretos) e ao observarmos o sistema no tempo t+1, vemos que estas características mudaram dentro do domínio de funcionamento deste modelo. A função F(P) indica as regras que determinam esta mudança.

**Modelagem de problemas no tempo e espaço.**

Na figura podemos ver ainda um modelo M no tempo t. Tal modelo mapeia conjuntos  características do problema P em características do modelo M, de modo que o modelo pode ser visto como uma simplificação da realidade. Pode-se dizer que a função E(P) é a função que mapeia as características do problema P nas características no modelo M. Além disso, temos a função f(M), que faz o papel da função F(P) para o modelo, determinando sua transição no tempo. O processo de modelagem consiste, portanto, em determinar E(P) e f(M).

As escolhas de E(P) e f(M) determinam o compromisso entre complexidade e confiabilidade do modelo. Espera-se que estas funções quando aplicadas ao problema, produzam modelos ao mesmo tempo simples e confiáveis, isto é, com um baixo erro embutido. Observe-se que na modelagem o erro é inevitável, dado que estamos priorizando certos aspectos da realidade em detrimento de outros, contudo a arte da modelagem consiste em selecionar exatamente aqueles aspectos que dão valor ao modelo.

## Como solucionar o modelo?

Dado que o modelo é definido, passa-se à solução. Nem todo modelo é passível de ser solucionado. Por exemplo, um modelo visual como um Diagrama de Casos de Uso não pode ser verificado por uma técnica qualitativa, ao contrário de um modelo de otimização. Portanto, deve-se entender que nesta seção nos referimos a este tipo de modelos.

Em primeiro lugar, antes de determinar qual a melhor forma de resolver o modelo, é importante entender qual a complexidade de solução do modelo. Alguns modelos exigem algoritmos custosos para serem solucionados. Isto não faz destes modelos melhores ou piores que outros, mas devemos ter em mente esta complexidade para evitarmos gastar tempo usando estratégias inadequadas.

Juntamente com a complexidade, devemos observar se o problema admite solução geral (solução analítica) ou se é possível apenas obter soluções para casos específicos (solução numérica). Este aspecto varia bastante de modelo para modelo e o conhecimento das diferentes técnicas de solução dos modelos ajudam o profissional a formar um cabedal de ferramentas para usar em cada caso. **Vide exemplo abaixo**.

Em seguida, é preciso avaliar a qualidade da solução observada, o que é feito, geralmente, por meio de técnicas relacionadas à forma de solução do modelo. A partir desta avaliação da qualidade da solução, pode-se determinar refinamentos no modelo, ou seja, caso  a solução não atinja a qualidade desejada, corrige-se o modelo introduzindo ou retirando características.

Por fim, como última etapa do processo de modelagem, deve-se interpretar o resultado. Encare esta etapa como um retorno aos termos e conceitos do problema definido. Na maior parte das vezes esta é uma etapa que o aprendiz dá pouca importância no processo de modelagem, contudo esta etapa é de suma importância porque é através dela que todo o esforço de modelagem e solução é coroado. É nesta etapa que os números obtidos na solução quantitativa são analisados tendo em vista a busca de uma resposta ao problema original.

### Tipos de solução

Considere o problema de achar o ponto de ótimo em uma função quadrática.

1. Qual a solução analítica?
2. Como encontrar uma solução numérica?

#### Solução analítica

```python
class quadratic:
    ''' implements a quadratic function f(x) = ax² + bx + c'''
    def __init__(self, a,b,c):
        self.a = a
        self.b = b
        self.c = c

    def solve(self, x):
        return self.a*x**2 + self.b*x + self.c

    def optimum(self):
        ''' finds the maximum or minimum point of this function
            through well-known formula '''
        delta = self.b**2 - 4*self.a*self.c
        return (- self.b/(2*self.a), -delta/(4*self.a))

    def __str__(self):
        return str(self.a)+"x² + "+str(self.b)+"x +"+str(self.c)
```

#### Solução Numérica

```python
def hillClimbing(f,ini,step,which):
    ''' hill climbing algorithm for finding the maximum or
        minimum value of a function

        f: a function to be solved
        ini: an initial point (float)
        step: the stepsize of the algorithm (float)
        which: a string 'min' or 'max'
        '''

    if which == 'min':
        def obj(x):
            ''' we always assume maximum'''
            return -f(x)
    elif which == 'max':
        def obj(x):
            return f(x)
    else:
        raise Exception()

    actual = ini
    right = ini + step
    left = ini - step

    while (not (obj(actual) > obj(right) and obj(actual) > obj(left) )):
        if obj(actual) > obj(right):
            actual = left
        elif obj(actual) < obj(right):
            actual = right
        right = actual + step
        left = actual - step

    return (actual,f(actual))
```

#### Exemplo

```python
f = quadratic(-2,1,10)
print("Equation: "+str(f))
(xopt,yopt) = f.optimum()
print("Optimum value through analytical solution x="+str(xopt)+" y="+str(yopt))
(xopt,yopt) = hillClimbing(f.solve,-40,0.001,'max')
print("Optimum value through numerical solution x="+str(xopt)+" y="+str(yopt))

f = quadratic(5,-10,20)
print("Equation: "+str(f))
(xopt,yopt) = f.optimum()
print("Optimum value through analytical solution x="+str(xopt)+" y="+str(yopt))
(xopt,yopt) = hillClimbing(f.solve,-40,0.001,'min')
print("Optimum value through numerical solution x="+str(xopt)+" y="+str(yopt))
```
