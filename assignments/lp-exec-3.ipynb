# Problem 1

The Petrisul oil refinery uses three types of crude oil (oil 1, oil 2, and oil 3) to produce three types of gasoline: regular, super, and extra.

To guarantee quality, each type of gasoline requires a composition of the various types of crude oil, as shown in the table below.

|Types of gasoline|Specs|
|--|--|
|Regular|No more than 70% of the oil 1|
|Super|No more than 50% of the oil 1|
||No less than 10% of the oil 2|
|Extra|No more than 50% of the oil 2|
||No less than 40% of the oil 3|

To meet customer demand, the refinery needs to produce at least 5,000 barrels per day of regular gasoline and 3,000 barrels per day of super and extra gasoline. The available daily capacity is 10,000 barrels of crude oil 1, 8,000 of crude oil 2, and 7,000 of crude oil 3. The refinery can produce up to 20,000 barrels of gasoline per day.

The refinery profits \$5 for each barrel of regular gasoline produced, \$7 per barrel of super gasoline, and $8 per barrel of extra gasoline. The costs of crude oil 1, 2, and 3 per barrel are \$2, \$3, and \$3, respectively. Formulate and solve the linear programming problem to maximize daily profit.
