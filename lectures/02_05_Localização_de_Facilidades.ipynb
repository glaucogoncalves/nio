#Problema de Localização de Facilidades

Imagine um problema de transporte clássico, mas onde vou decidir quais fornecedores devo manter funcionando.

Este é um problema de **programação binária mista**.

$C_{ij}$ = custo de transporte da localidade $j$ para o consumidor $i$

$F_j$ = custo fixo de manter a localidade $j$ aberta

$U_{j}$ = capacidade máxima da localidade $j$

$D_i$ = demanda do consumidor $i$

Variáveis de decisão:

$y_{ij}$ = quantidade transportada da localidade $j$ para o consumidor $i$

$x_{j}$ = 1 se a localidade $j$ for aberta; 0 caso contrário

###Exemplo

Uma empresa do setor alimentício está planejando abrir novas fábricas considerando cinco possibilidades: Manaus, Fortaleza, Vitória, Barueri e Curitiba. A partir das novas localidades, os produtos serão entregues para cinco consumidores finais: São Luís, Brasília, Belo Horizonte, Rio de Janeiro e São Paulo.

Cada fábrica possui um custo fixo de manutenção e uma capacidade máxima, conforme mostra a tabela abaixo. Os custos de transporte por unidade transportada de cada fábrica para cada consumidor final, além da demanda dos consumidores, também estão detalhados na tabela. A empresa quer determinar qual(is) fábrica(s) abrir, de modo a minimizar a soma dos custos fixos de instalação e dos custos de transporte, garantindo que a demanda dos consumidores finais será atendida.
