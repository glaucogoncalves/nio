## O que  um modelo de otimizao?

* Uma funo objetivo que deve ser maximizada ou minimizada, ex:
 * Minimizar o tempo de viagem entre Recife e Belm
 * Maximizar o nmero de clientes atendidos em uma loja
* Um conjunto de restries que devem ser cumpridas, ex:
 * No quero gastar mais do que $100
 * Devo estar em Belm s 20h
 * Um atendente no pode ficar mais de 20% do tempo ocioso

## Componentes de um problema de otimizao

**Variveis do problema**
* Inteiras, binrias, contnuas...

**Funo Objetivo**
* Funo sobre as variveis do problema (Linear / No-linear)
* Pode haver mais de uma (mono vs multi-objetivo)
* Minimizar ou Maximizar

**Restries**
* Inequaes sobre as variveis do problema
* Equaes sobre as variveis do problema

**Domnio das variveis do problema**
* Indicam o espao de busca sobre a varivel

# Modelagem de problemas de otimizao

## Forma geral de um problema de otimizao

O problema pode possui uma ou mais funes objetivo (representadas abaixo de forma vetorial)

$f(\pmb{x}) = [f_1(\pmb{x}),f_2(\pmb{x}),\ldots,f_m(\pmb{x})]$

Cada funo pode ter como fim minimizar ($\min$) ou maximizar ($\max$). Caso $m=1$, temos um problema monoobjetivo.

O problema ainda pode ser irrestrito ou restrito, i.e., estar sujeito  mltiplas restries, que podem ser equaes ou inequaes:

$h_i(\pmb{x}) = 0$, $(i=1,2,\ldots,M)$,

$g_j(\pmb{x}) \leq 0$, $(j=1,2,\ldots,N)$

Alm disso, cada varivel pertencente a um domnio:

$\pmb{x} = (x_1,x_2,\ldots,x_D) \in \mathbb{R}^D$

O domnio poderia ser ainda:
 * contnuo ou inteiro
 * limitado ou ilimitado

## Subtipos

* A rea de otimizao  chamada de programao matemtica
* Diversas subreas
  * Programao linear
    * Estuda problemas onde a funo objetivo e as restries so lineares
  * Programao inteira
    * Estuda problemas lineares com uma ou mais variveis inteiras
  * Otimizao combinatria
    * Estuda problemas lineares ou no lineares onde o espao de solues  discreto e limitado
  * Programao no-linear
    * Estuda o caso geral quando a funo objetivo ou as restries contm partes no-lineares

# Soluo de problemas de otimizao

* Como solucionar um problema de otimizao?
 * Depende da complexidade do problema
 * Depende dos tipos das variveis
 * Depende das funes
* Mtodos analticos
  * Sem restries
    * Mtodo da derivada
  * Com restries
    * Relaxao de Lagrange + mtodo da derivada
* Mtodos Numricos
  * Baseados na execuo de um algoritmo


## Mtodos Numricos (Algoritmos)

* Essncia de um Algoritmo de Otimizao
  * Gera uma soluo melhor $\pmb{x}^{(t+1)}$ a partir de uma soluo existente $\pmb{x}^{(t)}$, tal que $\pmb{x}^{(t+1)} = A(\pmb{x}^{(t)},\alpha)$, onde $\alpha$  um conjunto de parmetros e $A$  uma estratgia de atualizao da soluo usada pelo algoritmo
  * Algoritmos baseados em populaes usam mltiplos caminhos
  * Algoritmos diferentes $\Rightarrow$ formas diferentes de gerar solues
* Exatos (Convencionais)
  * Simplex (PL)
  * Busca Exaustiva (fora bruta)
  * Branch-and-Bound (PLI)
  * Mtodo de Newton (PNL)
* Heursticas: "seek <u>good feasible solutions</u> to optimization problems in circumstances where <u>the complexity of the problem</u> or the <u>limited time available</u> for its solution <u>do not allow exact solution</u>" [1]
  * Heursticas especficas para o problema em si
    * Aproximaes
  * Meta-heursticas
    * Algoritmos Genticos, Simulated Annealing, PSO...



[1] [Rardin, R.L. and Uzsoy, R., 2001. Experimental evaluation of heuristic optimization algorithms: A tutorial. Journal of Heuristics, 7, pp.261-304.](https://link.springer.com/article/10.1023/A:1011319115230)

### Algoritmos Convencionais vs Algoritmos Bio-inspirados

* Algoritmos convencionais
  * A otimalidade global (so **buscas locais**) depende do problema
    * A soluo de problemas lineares ou no-lineares convexos  global, mas outros problemas no possuem esta garantia
  * Os resultados dependem dos pontos iniciais (exceto para problemas lineares e convexos)
  * Tem dificuldade com discontinuidades na funo objetivo
* Algoritmos bio-inspirados
  * Aumentam as chances de encontrar solues globais
  * Resolvem uma grande classe de problemas (so como caixas pretas)
  * Podem ser potencialmente mais caros computacionalmente

# Exemplo

Certa empresa fabrica 2 produtos P1 e P2. O lucro por unidade de P1  de \$100 e o lucro unitrio de P2  de \$150. A empresa necessita de 2 horas para fabricar uma unidade de P1 e 3 horas para fabricar uma unidade de P2. O tempo mensal disponvel para essas atividades  de 120 horas. A demanda esperada para P1 por ms  de 40 unidades e para P2  de 30 unidades por ms. Quanto a empresa deveria fabricar de cada produto mensalmente de modo a maximizar seu lucro?

# Modelo

$P_1$  a quantidade de produtos do tipo P1 produzidos

$P_2$  a quantidade de produtos do tipo P2 produzidos

$
\max 100P_1 + 150P_2 \\
s.t. \\
2P_1 + 3P_2 \leq 120 \\
P_1 \leq 40 \\
P_2 \leq 30
$

#Soluo

## Instalando a biblioteca de modelagem

```python
!pip install -q pyomo
```

## Transcrevendo o modelo

```python
from pyomo.environ import *

# create a model
model = ConcreteModel()

# declare decision variables
model.p1 = Var(domain=NonNegativeIntegers)
model.p2 = Var(domain=NonNegativeIntegers)

# declare objective
model.lucro = Objective(expr = 100*model.p1 + 150*model.p2, sense=maximize)

# declare constraints
model.horas = Constraint(expr = 2*model.p1 + 3*model.p2 <= 120)
model.demandaP1 = Constraint(expr = model.p1 <= 40)
model.demandaP2 = Constraint(expr = model.p2 <= 30)

model.pprint()
```

## Instalando o solver

```python
!apt-get install -y -qq glpk-utils
```

## Solucionando

```python
SolverFactory('glpk', executable='/usr/bin/glpsol').solve(model).write()

# display solution
print('\nLucro = ', model.lucro())

print('\nDecision Variables')
print('P1 = ', model.p1())
print('P2 = ', model.p2())

print('\nConstraints')
print('Horas  = ', model.horas())
print('Demanda P1 = ', model.demandaP1())
print('Demanda P2 = ', model.demandaP2())
```
